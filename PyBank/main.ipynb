```python
# -*- coding: utf-8 -*-
"""
This script will use the Pathlib library to set the file path, use the csv library to read in the file, and iterate over each row of the file to get the monthly profit and Loss. 
Then it keeps tracks of the total months and calculates the net total profit or loss amount. It also keeps track of the change in profit loss for each month and then calculates 
the average of that and finds the maximum profit and minimum profit along with the month in which those maximum and minimum profits were recorded. It then Prints the analysis to system output 
and also exports the analysis to a text file that contains the final results
"""

# Import the pathlib and csv library
from pathlib import Path
import csv
import sys
```

```python
# Set the file path
csvpath = Path('./Resources/budget_data.csv')
```

```python
# Open the csv file as an object
with open(csvpath, 'r') as csvfile:
    # Initialize a variable to store the total number of months
    total_months = 0
    # Initialize a varaiable to store the total profit or loss amount
    net_total_profit_loss_amt = 0
    # Initialize a variable to store the average change across all months
    average_change = 0
    # Initialize a variable to keep track of profit/loss for each month that is read from the CSV file
    profit_loss = 0
    # Initialize a variable to keep track of profit/loss from the previous month so that the difference in profit/loss can be calculated
    profit_loss_prev = 0    
    # Initialize a variable to keep track of the month so that it can be used to add to a list that can be used in conjunction with a list that stores monthly change in profit/loss  
    month=''    
    # Initialize an array to store the profit/loss change for each month
    profit_loss_change_list = []
    # Initialize an array to store the months for profit change
    month_change_list = []
    # Initialize a varaiable that can be used to store the counts of profit/loss change array so that average can be calculated
    count_profit_loss_change = 0 
    # Initialize a varaiable to store the average profit or loss change across all months 
    average_profit_loss_change=0
    # Initialize a varaiable to store the total profit change across all months so that the average profit or loss change can be calculated
    total_profit_loss_change=0
    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    # Read the header row first (skip this step if there is no header)
    header = next(csvreader)
    # Read each row of data after the header
    for row in csvreader:
        # Increment the months as each rows are read  
        total_months+= 1        
        # Store the profit/loss from the previous month so that we can calculate the change
        profit_loss_prev = profit_loss
        # Store the profit/loss from the previous month so that we can calculate the change
        profit_loss = int(row[1])
        # Store the month for each row so that we can add that to a month array and use in conjunction with another array used to store the changes across each month
        month = row[0]
        # find the change in profit/loss for each month by subtracting the previous months profit/loss from current months profit/loss
        profit_loss_change = profit_loss - profit_loss_prev
        # check if the month value is 1 which denotes we are at the first row. We dont want to store the change as we only need to store the difference between current and previous starting from second row
        if (total_months!= 1):
            # add the change in profit/ loss to an array 
            profit_loss_change_list.append(profit_loss_change)
            # add the month to an array so that we can use that later while printing the month in which max or minimum profit/loss happened
            month_change_list.append(month)
        # calculate the total profi loss amount across all months by adding profit/loss values from each row 
        net_total_profit_loss_amt +=profit_loss
    # loop through the profit/loss change array so that we can calculate the total profit/loss and then use that to calculate the average value across all months
    for profit_loss_change in profit_loss_change_list:
        # add the counter to keep track of the number of items in the array to be used to calculate the average value across all months
        count_profit_loss_change+=1
        # accumulate the profit/loss for each month
        total_profit_loss_change+=profit_loss_change        
    # Calculate the average value of profit/loss across all months by dividing the total profit loss across each month by the number of months and then round that to 2 decimals
    average_profit_loss_change= round(total_profit_loss_change / count_profit_loss_change,2)

# Find the maximum value of profit loss across all months using the max function
max_value = max(profit_loss_change_list)
# Find the index value of the maximum value of profit loss so that we can use that to get the month in which the maximum profit loss happened. 
max_index = profit_loss_change_list.index(max_value)
# Find the month in which maximum profit loss happened
max_month = month_change_list[max_index]
# Find the minimum value of profit loss across all months using the min function 
min_value = min(profit_loss_change_list)
# Find the index value of the minimum value of profit loss so that we can use that to get the month in which the minimum profit loss happened. 
min_index = profit_loss_change_list.index(min_value)
# Find the month in which minimum profit loss happened
min_month = month_change_list[min_index]

# Print all the Analysis data calculated into the console/ standard output 
print('Financial Analysis')
print('----------------------------')
print(f"Total Months: {total_months}")
print(f"Total: ${net_total_profit_loss_amt}")
print(f"Average Change: ${average_profit_loss_change}")
print(f"Greatest Increase in Profits: {max_month} (${max_value})")
print(f"Greatest Decrease in Profits: {min_month} (${min_value})")

# Save a reference to the original standard output
original_stdout = sys.stdout 
# Open a file financial_analysis.txt in write mode so that we can Print the analysis into a text file that contains the final results
with open('financial_analysis.txt', 'w') as f:
    # Change the standard output to the file we created.
    sys.stdout = f 
    # Print all the Analysis data calculated into the text file
    print('Financial Analysis')
    print('----------------------------')
    print(f"Total Months: {total_months}")
    print(f"Total: ${net_total_profit_loss_amt}")
    print(f"Average Change: ${average_profit_loss_change}")
    print(f"Greatest Increase in Profits: {max_month} (${max_value})")
    print(f"Greatest Decrease in Profits: {min_month} (${min_value})")
    # Reset the standard output to its original value
    sys.stdout = original_stdout 
```
